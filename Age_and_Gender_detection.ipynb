```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Load your dataset
df = pd.read_csv('age_gender.csv')

# Preprocess the data
X = np.array(df['pixels'].str.split(' ', expand=True).astype(float))
y_age = np.array(df['age'])
y_gender = np.array(df['gender'])

# Normalize the pixel values
X = X / 255.0

# Split the data into training and testing sets
X_train, X_test, y_age_train, y_age_test, y_gender_train, y_gender_test = train_test_split(
    X, y_age, y_gender, test_size=0.2, random_state=42)

# Create a model for age detection
age_model = Sequential()
age_model.add(Dense(128, input_shape=(X_train.shape[1],), activation='relu'))
age_model.add(Dense(64, activation='relu'))
age_model.add(Dense(1, activation='linear'))

# Compile the model
age_model.compile(optimizer='adam', loss='mean_squared_error')

# Train the age model
age_model.fit(X_train, y_age_train, epochs=10, batch_size=32, validation_data=(X_test, y_age_test))

# Create a model for gender detection
gender_model = Sequential()
gender_model.add(Dense(128, input_shape=(X_train.shape[1],), activation='relu'))
gender_model.add(Dense(64, activation='relu'))
gender_model.add(Dense(1, activation='sigmoid'))

# Compile the model
gender_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Encode gender labels (0 for male, 1 for female)
label_encoder = LabelEncoder()
y_gender_train_encoded = label_encoder.fit_transform(y_gender_train)
y_gender_test_encoded = label_encoder.transform(y_gender_test)

# Train the gender model
gender_model.fit(X_train, y_gender_train_encoded, epochs=10, batch_size=32, validation_data=(X_test, y_gender_test_encoded))

# Evaluate the models
age_predictions = age_model.predict(X_test)
gender_predictions = gender_model.predict(X_test)

# Decode gender predictions
gender_predictions_decoded = label_encoder.inverse_transform((gender_predictions > 0.5).astype(int))

# Print classification report for gender detection
print("Classification Report for Gender Detection:")
print(classification_report(y_gender_test, gender_predictions_decoded))

# You can use age_predictions and gender_predictions for your age and gender detection system.

```
